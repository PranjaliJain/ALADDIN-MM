```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cols = ['unit','cycle','avg_power','avg_fu_power','total_area','fu_area']
mat2 = ['2x2', 14, 2.25627, 0.690865, 70534.9, 7657.49]
mat4 = ['4x4', 102, 2.35191, 0.775197, 71109.1, 8231.71 ]
mat8 = ['8x8', 848, 2.29468, 0.753148, 71300.5, 8423.12]
mat16 =['16x16',6942, 2.26559, 0.742529, 71683.4, 8805.93 ]
mat32 =['32x32', 56486, 2.24226, 0.730509, 71683.4, 8805.93]
```

```python
mat = [mat2, mat4, mat8, mat16, mat32]
```

```python
df = pd.DataFrame(mat, columns=cols)
```

```python
df['cycle'][2]
```

```python
mat[2][0]
```

```python
# plt.scatter(df['cycle'],df['avg_power'])
fig, ax = plt.subplots()
ax.scatter(df['cycle'],df['avg_power'])

for i,_ in enumerate((mat)):
    ax.annotate(df['unit'][i], (df['cycle'][i],df['avg_power'][i]))
plt.ylabel("Average Power(mW)")
plt.xlabel("Number of cycles")
plt.show()
```

```python
# plt.scatter(df['cycle'],df['fu_area'])
fig, ax = plt.subplots()
ax.scatter(df['cycle'],df['fu_area'])

for i,_ in enumerate((mat)):
    ax.annotate(df['unit'][i], (df['cycle'][i],df['fu_area'][i]))
plt.ylabel("FU Area (uM^2)")
plt.xlabel("Number of cycles")
plt.show()
```

```python
# plt.scatter(df['cycle'],df['fu_area'])
fig, ax = plt.subplots()
ax.scatter(df['fu_area'],df['avg_power'])

for i,_ in enumerate((mat)):
    ax.annotate(df['unit'][i], (df['fu_area'][i], df['avg_power'][i]))
plt.xlabel("FU Area (uM^2)")
plt.ylabel("Average Power(mW)")
plt.show()
```
